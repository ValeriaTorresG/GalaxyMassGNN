```python
def find_negative_indices(df):
    indices = np.union1d(np.where(df.FLUX_W1_DERED < 0)[0], np.where(df.FLUX_W2_DERED < 0)[0])
    return indices

dic = {}
for r in [3, 6, 7, 11, 12, 13, 14]:
    df = pd.read_csv(f'./data/rosette{r}_nodes.csv')
    negative_indices = find_negative_indices(df)
    dic[r] = negative_indices

dic
```
