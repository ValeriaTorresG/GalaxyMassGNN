```python
import numpy as np
from astropy.io import fits

file = fits.open('./data/frame-r-005071-3-0347.fits')
file
```

```python
file.info()
```

```python
d = file[0].data
np.shape(d)
```

```python
import matplotlib.pyplot as plt

plt.imshow(d, cmap='jet')
plt.colorbar()
```

```python
%matplotlib widget

plt.imshow(d, cmap='jet')
plt.colorbar()
```

### Escala log

```python
from matplotlib.colors import LogNorm

plt.figure()
plt.imshow(d, cmap='jet', norm=LogNorm())
```

### Arregla origenes de los ejes

```python
plt.figure()
plt.imshow(d, cmap='jet', norm=LogNorm(), origin='lower')
plt.colorbar(location='top')
```
